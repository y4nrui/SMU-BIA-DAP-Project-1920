```python
import pandas as pd
import gender_guesser.detector as gender
import re
import time
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/OverallData_WithGender.csv')
```

```python
data['gender'].value_counts()
```

```python
def gender_visualisation(data):
    """ 
    function to generate visualisation to see which gender books more
    input: dataframe
    """
    data_temp = data
    data_tempf = data_temp[(data_temp['gender'] == 'female')]
    data_tempm = data_temp[(data_temp['gender'] == 'male')]
    data_temp = pd.concat([data_tempf, data_tempm]) 
    sns.countplot(x='gender', data=data_temp, palette='Set3')
```

```python
gender_visualisation(data)
```

```python
def gender_visualisation_all(data):
    """ 
    function to generate visualisation to see which gender books more
    input: dataframe
    """
    data_temp = data
    sns.countplot(x='gender', data=data_temp, palette='Set3')
```

```python
gender_visualisation_all(data)
```
